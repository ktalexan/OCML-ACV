## <font color='green'>Geoprocessing operations</font>

Load and create an object instance of the class for the area with the project's directories and attributes.

```python
azg = acvgis(ntustinList, area, prjdata, prjagp, agpname)
```

```python
azg.CreateDomains()
```

```python
azg.CreateDatasets()
```

```python
azg.AddGeoJSON()
```

```python
azg.CreateCardinalFeatures()
```

```python
azg.MergeAreas()
```

```python
azg.ProcessDetections()
```

```python
responseStopSigns = azg.CreatePointClusters('StopSigns', 'NorthTustinCardinals1StopSigns', 50)
```

```python
responseFireHydrants = azg.CreatePointClusters('FireHydrants', 'NorthTustinCardinals1FireHydrants', 50)
```

```python
azg.CreateObjectFeatureClass('NorthTustinCardinals1StopSignsFPC', responseStopSigns)
```

```python
azg.CreateObjectFeatureClass('NorthTustinCardinals1FireHydrantsFPC', responseFireHydrants)
```
